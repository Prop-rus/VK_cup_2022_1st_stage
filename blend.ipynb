```python
import pandas as pd
```

```python
bert = pd.read_csv('./output/bert_output.csv')
bert
```

```python
bert_clean = pd.read_csv('./output/bert_clean_output.csv')
bert_clean
```

```python
lstm = pd.read_csv('./output/lstm.csv')
lstm
```

```python
blend = pd.concat((
    bert[['Unnamed: 0', 'oid', 'category', 'prob']],
    
    bert_clean[['Unnamed: 0', 'oid', 'category', 'prob']],
    
    lstm[['Unnamed: 0', 'oid', 'category', 'prob']]
))
blend
```

```python
blend[blend['Unnamed: 0']==0]
```

```python
blend_fin = blend.sort_values(['Unnamed: 0', 'prob'], ascending=[True, False])
blend_fin
```

```python
blend_fin = blend_fin.groupby(['Unnamed: 0']).head(1)
blend_fin
```

```python
test = pd.read_csv('./test.csv')
test
```

```python
blend_fin = pd.concat((blend_fin.reset_index(), test), axis=1)
```

```python
blend_fin = blend_fin.loc[:,~blend_fin.columns.duplicated()]
blend_fin
```

```python
blend_fin[blend_fin.prob > 0.9999][['oid','category', 'text']].to_csv('./data/blended_test.csv', index=False)
```
